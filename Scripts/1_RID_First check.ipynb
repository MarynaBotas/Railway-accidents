## 5. Data Consistency Check

### Duplicates

```python
# Identify duplicate rows
duplicates = df_selected_1.duplicated()
```

```python
# Display the duplicate rows
duplicate_rows = df_selected_1[duplicates]

print("\nDuplicate rows:")
print(duplicate_rows)
```

```python
df_no_duplicates = df_selected_1.drop_duplicates()
```

```python
df_no_duplicates.shape
```

### Missing values

```python
#Looking for missing values
df_no_duplicates.isnull().sum()
```

```python
df_no_duplicates.dtypes
```

### Addressing Missing Values

#### The 'latitude' and 'longitude' columns have too many missing values, making them useless for the further analysis.

```python
# Drop columns 'latitude' and 'longitude'
df_no_duplicates = df_no_duplicates.drop(columns=['latitude', 'longitude'])
```

#### Columns 'visibility', 'weather_condition' and 'has_passengers' have object datatypes. Therefore, we replace missing values with common category 'Unknown' for these variables.

```python
# Fil miising values of the variables 'visibility', 'weather_condition' and 'has_passengers' with a common category 'Unknown'
df_no_duplicates['visibility'].fillna('Unknown', inplace=True)
df_no_duplicates['weather_condition'].fillna('Unknown', inplace=True)
df_no_duplicates['has_passengers'].fillna('Unknown', inplace=True)
```

#### Other columns have 1 to 3 missing values, and dropping them in a data set with 210672 rows should not skew the data any more than any other way of addressing.

```python
# Drop rows with any missing values
df_cleaned_na = df_no_duplicates.dropna()
```

```python
# Ckeck the output
df_cleaned_na.shape
```

```python
df_cleaned_na.head()
```

### Mixed data types

```python
# Check for mixed types
for col in df_cleaned_na.columns.tolist():
    weird = (df_cleaned_na[[col]].map(type)!=df_cleaned_na[[col]].iloc[0].apply(type)).any(axis=1)
    if len (df_cleaned_na[weird])>0:
        print (col)
```

```python
# Check actual data types
df_cleaned_na.dtypes
```

```python
# Convert 'year', 'month', and 'fra_district' columns to integer
df_cleaned_na['year'] = df_cleaned_na['year'].astype(int)
df_cleaned_na['month'] = df_cleaned_na['month'].astype(int)
df_cleaned_na['fra_district'] = df_cleaned_na['fra_district'].astype(int)
```

```python
# Convert 'year' column to integers using .loc
df_cleaned_na.loc[:, 'year'] = pd.to_numeric(df_cleaned_na['year'], errors='coerce').fillna(0).astype(int)

# Convert 'month' column to integers using .loc
df_cleaned_na.loc[:, 'month'] = pd.to_numeric(df_cleaned_na['month'], errors='coerce').fillna(0).astype(int)

# Convert 'fra_district' column to integers using .loc
df_cleaned_na.loc[:, 'fra_district'] = pd.to_numeric(df_cleaned_na['fra_district'], errors='coerce').fillna(0).astype(int)
```

```python
df_cleaned_na.loc[:, 'damage_cost'] = pd.to_numeric(df_cleaned_na['damage_cost'], errors='coerce').fillna(0).astype(float)
```

```python
# Check the outcome
df_cleaned_na.dtypes
```

```python
# Identify non-numeric values
non_numeric = df_cleaned_na[~df_cleaned_na['damage_cost'].apply(lambda x: str(x).replace('.', '', 1).isdigit())]
print("Non-numeric values in 'damage_cost' column:")
print(non_numeric)
```

```python
# Convert 'damage_cost' to numeric, filling NaNs with 0 if necessary
df_cleaned_na['damage_cost'] = pd.to_numeric(df_cleaned_na['damage_cost'], errors='coerce')
```

```python
# Using .loc to ensure setting the value on the original DataFrame
df_cleaned_na.loc[:, 'damage_cost'] = pd.to_numeric(df_cleaned_na['damage_cost'], errors='coerce')
```

```python
# Fill NaNs with 0 if needed
df_cleaned_na.loc[:, 'damage_cost'] = df_cleaned_na['damage_cost'].fillna(0).astype(float)
```

# 6. Data Profile

```python
# Summary statistics
df_cleaned_na.describe().astype(int)
```

```python
# Value counts for 'temperature' 
df_cleaned_na['temperature'].value_counts(dropna = False)
```

```python
# Calculate the value counts and sort them from lowest value to highest
temperature_counts_sorted = df_cleaned_na['temperature'].value_counts(dropna=False).sort_index()
```

```python
# Check the output
print(temperature_counts_sorted)
```

```python
# Identifying Outliers
outliers = df_cleaned_na[df_cleaned_na['temperature'] > 134]
outliers.shape[0]
```

```python
# Calculate the mean of the non-outlier values
mean_temperature = df_cleaned_na[df_cleaned_na['temperature'] <= 134]['temperature'].mean()
```

```python
# Replace outliers with the mean
df_cleaned_na.loc[df_cleaned_na['temperature'] > 134, 'temperature'] = mean_temperature
```

```python
# Check the output
temperature_counts_sorted = df_cleaned_na['temperature'].value_counts(dropna=False).sort_index()
print(temperature_counts_sorted)
```

```python
# Summary statistics after changes made
df_cleaned_na.describe().astype(int)
```

```python
# Check unique values of categorical data
for column in df_cleaned_na.select_dtypes(include='object').columns:
    unique_values = df_cleaned_na[column].unique()
    print(f"\nUnique values in {column}:")
    print(unique_values)
```

```python
# Mode and cardinality for each categorical column
modes = df_cleaned_na.mode().iloc[0]
cardinality = df_cleaned_na.nunique()
print("\nMode and Cardinality:")
for column in df_cleaned_na.select_dtypes(include='object').columns:
    print(f"{column}: Mode = {modes[column]}, Cardinality = {cardinality[column]}")
```
